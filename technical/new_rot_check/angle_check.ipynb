```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('../baseline/baseline_v2.1_10yrs.db')
baseline = pd.read_sql('select rotTelPos,rotSkyPos,note from observations;', conn)
conn.close()
```

```python
conn = sqlite3.connect('new_rot_check_v2.1_10yrs.db')
new_rot = pd.read_sql('select rotTelPos,rotSkyPos,note from observations;', conn)
conn.close()
```

```python
_ack = plt.hist(baseline['rotTelPos'], bins=100)
plt.xlabel('rotTelPos (deg)')
_ack = plt.hist(new_rot['rotTelPos'], bins=100, alpha=0.5)
```

```python
_ack = plt.hist(baseline['rotSkyPos'], bins=100)
plt.xlabel('rotSkyPos (deg)')
_ack = plt.hist(new_rot['rotSkyPos'] % 360, bins=100, alpha=0.5)
```

```python
ddf_names = [name for name in np.unique(baseline['note']) if 'DD' in name]
```

```python
for ddf in ddf_names:
    plt.figure()
    good = np.where(baseline['note'] == ddf)
    _ack = plt.hist(baseline['rotSkyPos'].values[good], bins=100)
    
    good = np.where(new_rot['note'] == ddf)
    _ack = plt.hist(new_rot['rotSkyPos'].values[good] % 360, bins=100, alpha=0.5)
    plt.xlabel('RotSkyPos (deg)')
    plt.title(ddf)
```
